## Data Merging

```python
import numpy as np
import pandas as pd

df1={

    'id':[1,2,3],
    'name':["ahmed","fatima","Ali"],
    'age':[28,32,24]
}
d1=pd.DataFrame(df1)

df2={
    'id':[1,2,4],
    'city':["karachi","islamabad","lahore"],
    'occupation':["teacher","doctor","engr"]
}
d2=pd.DataFrame(df2)
d2
```

```python
merged_df=pd.merge(d1,d2, on="id",how='inner')
merged_df
```

## DAta Concatenation

```python
import pandas as pd

data1={'Name':['Ahmed','Ali','Fatima','Guluna','Rubab'],'age':[23,25,36,65,32]}

data2={'Name':['saba','dada','naba','amber','adeel'],'age':[23,54,23,34,54]}

data3={'City':['Karachi','Lahore','Islamabad','Punjab','Sakhar'],'Provinces':['Sindh','Punjab','KPK','Balochistan','Sarhad']}

df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
df3=pd.DataFrame(data3)

concat_df=pd.concat([df1,df2,df3])
concat_df
## rows and colums m b kar sakhty 
## if only only rows add axis=0 attribute
## if only only columns add axis=1 attribute
```
